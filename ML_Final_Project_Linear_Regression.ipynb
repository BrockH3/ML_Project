```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
data = yf.Ticker("ETH-USD").history(period='5y')
ordered_data = data.sort_values(by=['Date'])
```

```python
varlist =  ['Open', 'High', 'Low','Volume']
X = ordered_data[varlist]

Y = ordered_data['Close']
```

```python

split_percentage = 0.8
split = int(split_percentage*len(data))

# Train data set
X_train = X[:split]
Y_train = Y[:split]

# Test data set
X_test = X[split:]
Y_test = Y[split:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model_Linear = LinearRegression()
model_Linear.fit(X_train, Y_train)

Y_Predicitions = model_Linear.predict(X_test)

mse = mean_squared_error(Y_test, Y_Predicitions)
print(f'MSE: {mse}')

plt.scatter(ordered_data.index, ordered_data['Close'], color='red', label='Close Prices')

plt.plot(X_test.index, Y_Predicitions, color='blue', linewidth=3, label='Predicted Close Prices')

plt.title('Ethereum Close Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
